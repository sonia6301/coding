```python
import pandas as pd
import matplotlib.pyplot as plt

# URL of the CSV file
url = 'https://raw.githubusercontent.com/sonia6301/coding/main/data/NWT_04_Clean_Data.csv'

# Read the CSV file directly from the URL
df = pd.read_csv(url)

# Parse the 'Standard Time' column as datetime
df['Standard Time (24 Hour) YYYY/MM/DD HH:MM:SS'] = pd.to_datetime(df['Standard Time (24 Hour) YYYY/MM/DD HH:MM:SS'])

# Set the 'Standard Time' as the index
df.set_index('Standard Time (24 Hour) YYYY/MM/DD HH:MM:SS', inplace=True)

# Create a figure and a set of subplots
fig, axs = plt.subplots(3, 1, figsize=(14, 15), sharex=True)

# Plot 'Depth (m)' on the first subplot
axs[0].plot(df.index, df['Depth (m)'], label='Depth (m)', color='blue')
axs[0].set_ylabel('Depth (m)')
axs[0].set_title('Time Series of Depth (m)')
axs[0].legend()
axs[0].grid(True)

# Plot 'Pressure (psia)' on the second subplot
axs[1].plot(df.index, df['Pressure (psia)'], label='Pressure (psia)', color='red')
axs[1].set_ylabel('Pressure (psia)')
axs[1].set_title('Time Series of Pressure (psia)')
axs[1].legend()
axs[1].grid(True)

# Plot 'Depth (m)' and 'Pressure (psia)' on the third subplot
axs[2].plot(df.index, df['Depth (m)'], label='Depth (m)', color='blue')
axs[2].plot(df.index, df['Pressure (psia)'], label='Pressure (psia)', color='red')
axs[2].set_xlabel('Time')
axs[2].set_ylabel('Measurements (m and psia)')
axs[2].set_title('Time Series of Depth (m) and Pressure (psia)')
axs[2].legend()
axs[2].grid(True)

# Adjust the layout
plt.tight_layout()
plt.show()
```
