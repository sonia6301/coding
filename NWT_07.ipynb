```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV data from the URL, skipping the first row of actual data
url = "https://raw.githubusercontent.com/sonia6301/coding/main/data/NWT_07.csv"
data = pd.read_csv(url)

# Inspect the column names
print(data.columns)

# Rename columns for easier access (if necessary)
data.columns = data.columns.str.strip()  # Strip any leading/trailing whitespace
data.columns = data.columns.str.replace(' ', '_')  # Replace spaces with underscores

# Convert the 'Standard Time (24 Hour)' column to datetime
data['Standard_Time_(24_Hour)'] = pd.to_datetime(data['Standard_Time_(24_Hour)'], errors='coerce')

# Plot Depth over time
plt.figure(figsize=(10, 6))
plt.plot(data['Standard_Time_(24_Hour)'], data['Depth'], label='Depth', color='blue')
plt.xlabel('Time')
plt.ylabel('Depth (m)')
plt.title('Depth over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot psia over time
plt.figure(figsize=(10, 6))
plt.plot(data['Standard_Time_(24_Hour)'], data['Pressure'], label='Pressure (psia)', color='red')
plt.xlabel('Time')
plt.ylabel('Pressure (psia)')
plt.title('Pressure over Time')
plt.legend()
plt.grid(True)
plt.show()

# Plot Depth and psia together on the same figure
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Time')
ax1.set_ylabel('Depth (m)', color='blue')
ax1.plot(data['Standard_Time_(24_Hour)'], data['Depth'], label='Depth', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Pressure (psia)', color='red')
ax2.plot(data['Standard_Time_(24_Hour)'], data['Pressure'], label='Pressure (psia)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

fig.tight_layout()
plt.title('Depth and Pressure over Time')
plt.grid(True)
plt.show()
```
